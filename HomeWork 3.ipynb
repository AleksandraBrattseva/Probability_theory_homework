Задача 1
Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean): 
а) среднее арифметическое
б) среднее квадратичное отклонение
в) смещенную и несмещенную оценки дисперсий для данной выборки

```python
array = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
array
```

а) среднее арифметическое:

```python
def mean_value(array):
    return sum(array) / len(array)
print(f'Среднее арифметическое = {mean_value(array): .2f}')
```

б) среднее квадратичное отклонение:

```python
def mean_square_deviation(array):
    square_dev = (array - mean_value(array)) **2
    return (sum(square_dev) / len(square_dev)) **(1/2)
print(f'Среднее квадратичное отклонение = {mean_square_deviation(array): .2f}')
```

Задача 2:
В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
def combinations(n,k):
    return factorial(n) / (factorial(k) * factorial(n - k))
P1 = (combinations(3,2) / combinations(8,2)) * (combinations(5,3) * combinations(7,1) / combinations(12,4))
print(f'Вероятность события "из первой корзины не вытащили ни одного белого мяча'
      f', из второй вытащили 3 белых мяча"\nP(A1) = {P1: .2f}')
P2 = (combinations(5,1)*combinations(3,1)*combinations(5,2)*combinations(7,2))/(combinations(8,2)*combinations(12,4))
print(f'Вероятность события "из первой корзины вытащили 1 белый мяч'
      f', из второй вытащили 2 белых мяча"\nP(A2) = {P2: .2f}')
P3 = (combinations(5,2) * combinations(5,1) * combinations(7,3)) / (combinations(8,2) * combinations(12,4))
print(f'Вероятность события "из первой корзины вытащили 2 белых мяча'
      f', из второй вытащили 1 белый мяч"\nP(A2) = {P3: .2f}')
print(f'Вероятность того, что из вытащенных мячей 3 белые Р(А) = {P1 + P2 + P3: .2f}')    
```

Задача 3:
На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: 
a) первым спортсменом 
б) вторым спортсменом 
в) третьим спортсменом

```python
PB = 1/3
PA = PB * 0.9 + PB * 0.8 + PB * 0.6
print(f'Полная вероятность наступления события А Р(А) = {PA: .2f}')
PAB1 = PB * 0.9 / PA
PAB2 = PB * 0.8 / PA
PAB3 = PB * 0.6 / PA
print(f'Вероятность того, что выстрел произвёл первый спортсмен: {PAB1: .2f}\n'
      f'Вероятность того, что выстрел произвёл второй спортсмен: {PAB2: .2f}\n'
      f'Вероятность того, что выстрел произвёл третий спортсмен: {PAB3: .2f}')
```

Задача 4:
В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: 
a) на факультете A 
б) на факультете B 
в) на факультете C

```python
P = 0.25 * 0.8 + 0.25 * 0.7 + 0.5 * 0.9
print(f'Полная вероятность наступления события = {P}')
PA = 0.25 * 0.8 / P
PB = 0.25 * 0.7 / P
PC = 0.5 * 0.9 / P
print(f'Вероятность того, что студент учится на факультете А: {PA: .2f}\n'
      f'Вероятность того, что студент учится на факультете B: {PB: .2f}\n'
      f'Вероятность того, что студент учится на факультете C: {PC: .2f}')
```
